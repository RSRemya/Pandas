```python
import pandas as pd
```

# Merge Operation

```python
d1 = {'key1':[1,2,3,4,5,"sudh"],
      'key2' : [7,8,9,10,11,12],
      'key3':[13,14,15,16,17,18],
      'key4':[19,20,21,22,23,24]
}
```

```python
d2 = {'key1':[1,2,3,4,5,"sudh"],
      'key2' : [7,8,"kumar",10,11,12],
      'key3':[13,14,15,16,17,18],
      'key4':[19,20,21,"ineuron","pwskills",24]
}
```

```python
#converting dictionary into dataframe:
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)# by default they are joined by inner join, where only common data is included in new dataset so formed
```

```python
pd.merge(df1,df2,how= 'inner')# shows same result as above
```

```python
pd.merge(df1,df2,how= 'outer')#combines both df1 and df2 one after another combining similar rows as a single one
```

```python
pd.merge(df1,df2,how= 'inner',on= 'key1')#brings all data where entities of key1 is same for both dataframes, and keeps column all other columns from both df as separate ones
```

```python
pd.merge(df1,df2,how= 'inner',on= 'key4')#brings all data where entities of key4 is same for both dataframes, and keeps column all other columns from both df as separate ones
```

```python
pd.merge(df1,df2,how= 'left',on= 'key4')#shows all columns of df1(which is at the left), entries in df at right is NaN where any one row element is mismatching with df at left
```

```python
pd.merge(df1,df2,how= 'right',on= 'key4')
```

```python
pd.merge(df1,df2,how= 'inner',on= 'key4')
```

```python
pd.merge(df1,df2,how= 'outer',on= 'key4')# it will bring every combination from leftand right df
```

```python
pd.merge(df1,df2,how= 'outer',on= ['key4','key1'])
```

```python
# Note : NaN means not a number
```

```python
pd.merge(df1,df2,how= 'inner',on= 'key1')
```

```python
d11 = {'a':[1,2,3,4,5,"sudh"],
      'b' : [7,8,9,10,11,12],
      'c':[13,14,15,16,17,18],
      'd':[19,20,21,22,23,24]
}

df11 = pd.DataFrame(d11)

df11
```

```python
d12 = {'key1':[1,2,3,4,5,"sudh"],
      'key2' : [7,8,"kumar",10,11,12],
      'key3':[13,14,15,16,17,18],
      'key4':[19,20,21,"ineuron","pwskills",24]
}

df12 = pd.DataFrame(d12)
df12
```

```python
pd.merge(df11, df12 , how = 'left' , left_on = 'a' , right_on='key1')# when column headers are diff in both df, put column name of left dataset in left_on and column name of right dataset in right_on
```

# Join Operation

```python
d10 = {'key11':[1,2,3,4,5,"sudh"],
      'key21' : [7,8,9,10,11,12],
      'key31':[13,14,15,16,17,18],
      'key41':[19,20,21,22,23,24]
}
df10 = pd.DataFrame(d10 , index = [1,2,3,4,5,6])
df10
```

```python
d20 = {'key1':[1,2,3,4,5,"sudh"],
      'key2' : [7,8,"kumar",10,11,12],
      'key3':[13,14,15,16,17,18],
      'key4':[19,20,21,"ineuron","pwskills",24]
}
df20 = pd.DataFrame(d20, index=[1,2,3,4,5,6])
df20
```

```python
df10.join(df20)
```

```python
df10.join(df20,how = 'right')
```

```python
df10.join(df20,how = 'left')
```

```python
df10.join(df20 , how = 'outer')
```

```python
df10.join(df20 , how = 'inner')
```

# Concatenating

```python
pd.concat([df1,df2] , axis=0)# axis=0 means concatinating row-wise, by default axis is 0
```

```python
pd.concat([df1,df2] , axis=1)# axis=1 means concatinating column-wise
```
